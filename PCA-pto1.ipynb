## Algoritmo Covarianza

```python
#Algoritmo que devuelve la matriz de covarianza en base a un conjunto de datos presentado en una matriz(DataFrame)
def matrizCovarianza(data):
    #el tamaño de la muestra
    tam = len(TasasHom.index)
    #la matriz de promedios
    dim = len(TasasHom.T.index)
    #la dimension de cada elemento de la muestra
    cov = np.zeros((dim, dim))
    #promedio vectorial
    mean = data.mean(axis = 0)
    #variables de iteración
    i = j = k = 0
    #itera sobre una dimension de la matriz
    for i in range(dim):
        #itera sobre la otra dimension de la matriz, hasta la diagonal
        #(aprovechando la simetría de la matriz de covarianza)
        for j in range(0,i+1):
            suma = 0
            #Calcula la sumatoria que determina el término i,j de la matriz
            for k in range(tam):
                a = (data.ix[k,i] - mean[i])*(data.ix[k,j] - mean[j])
                suma += a
            #asigna los valores a los cuadros diagonalmente simétricos de la matriz de covarianza    
            cov[j][i] = suma/(tam-1)
            cov[i][j] = suma/(tam-1)

    return cov
```

## Matriz de covarianza del problema

```python
#matriz de covarianza de los datos de tasas de homicidio de los departamentos de Colombia a través de los años
cova = matrizCovarianza(TasasHom)
```

## Cálculo de autovalores y autovectores

```python
#saca los autovalores y autovectores de la matriz de covarianza
eig = np.linalg.eig(cova)
#valores propios con un numero asociado que indica la posición en la lista su autovector correspondiente
eigenValuesN = np.array([np.array([(eig[0])[k] , k]) for k in range(len(eig[0]))])
#Se ordena el array respecto a la columna de autovalores (de mayor a menor)
eigenValuesN = -eigenValuesN
eigenValuesN = -eigenValuesN[[eigenValuesN[:,0].argsort()]]
```
