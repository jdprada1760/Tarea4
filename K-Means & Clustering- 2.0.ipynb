```python
def clustering( nClusters, data ):
    i = j = 0
    dim = len(data[0])
    muestra = len(data)
    indexClusters = -np.ones(muestra)
    clusterPoints = 10*np.random.random((nClusters, dim))
    #clusterPoints = np.array([[0,0],[100,100]])
    termino = False
    while termino == False :
        hayClustersVacios = False
        indexClusters = -np.ones(muestra)
        i = 0
        for i in range(muestra):
            dmin = np.linalg.norm(data[i] - clusterPoints[0])
            nmin = 0
            for j in range(1, nClusters):
                n = np.linalg.norm(data[i] - clusterPoints[j])
                if n < dmin:
                    dmin = n
                    nmin = j
            indexClusters[i] = nmin
        
        newClusters = np.zeros((nClusters,dim))
        k = 0
        for k in range(nClusters):
            index = where(indexClusters == k)
            if len(index[0]) != 0 :
                newClusters[k] = np.mean(data[index], linspace(0,dim,1))
            else: 
                hayClustersVacios = True
        if (newClusters == clusterPoints).all():
            #c = where((newClusters == [0.,0.]).all())
            #print "\n", newClusters, c
            #if len(c[0]) != 0:
             #   newClusters = 10*np.random.random((nClusters, dim))
            #else:
            if hayClustersVacios == True:
                newClusters = 10*np.random.random((nClusters, dim))
            else:    
                termino = True
        #print indexClusters
        #print clusterPoints
        clusterPoints = newClusters
    k = 0       
    return [[data[where(indexClusters == k)] for k in range(nClusters)], clusterPoints]
        
```
